```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel("dados_matriculas.xlsx")

display(df)
```

```python
# Gráfico de matriculas mensais

plt.figure(figsize=(10, 5))
df['Matrículas'].plot(marker='o', linestyle='-', color='#328ba8')
plt.title('Matriculas Mensais')
plt.xlabel('Mês')
plt.ylabel('Número de matrículas')
plt.grid(True, color='#c9ad47')

plt.show()
```

```python
# Identificar o mês com maior e menor número de matrículas
maior_matricula = df['Matrículas'].max()
indice_maior_matricula = df['Matrículas'].idxmax()

print(indice_maior_matricula)
print(maior_matricula)
```

```python
menor_matricula = df['Matrículas'].min()
indice_menor_matricula = df['Matrículas'].idxmin()

print(indice_menor_matricula)
print(menor_matricula)
```

```python
plt.figure(figsize=(10, 5))
df['Matrículas'].plot(marker='x', mfc='#594fe8', mec='#275c4f', linestyle='--', color='#9e3a69', label='Matrículas')

# Informações de destaque
plt.axhline(y=maior_matricula, color='#3a7a9e', linestyle=':', linewidth=2, label=f"Máximo: {maior_matricula}")
plt.axhline(y=menor_matricula, color='#3a7a9e', linestyle=':', linewidth=2, label=f"Mínimo: {menor_matricula}")

plt.title('Matrículas Mensais')
plt.xlabel('Mês')
plt.ylabel('Número Matrículas')
plt.grid(True, color='gold')
plt.legend()
plt.show()
```
